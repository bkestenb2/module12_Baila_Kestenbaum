### F1. Shopping List Manager (1D list + functions) (Suggested)
Write a program that maintains a shopping list (a 1D list of strings).

**Required functions:**
- `add_item(items, item)`
- `remove_item(items, item)` (remove by value if present)
- `print_items(items)` (numbered list)
- `save_items(items, filename)` (one item per line)
- `load_items(filename)` (returns a list)

**Program behavior:**
Provide a simple text menu in a loop:
1) add  2) remove  3) view  4) save  5) load  6) quit
The list should update as the user makes choices.


````python
# TODO: Implement F1 here
def add_items(items, item):
  items.append(item)

def remove_items(items,item):
  items.remove(item)

def print_items(items):
  if len(items) == 0:
    print("List is empty")
  else:
    for i in range(len(items)):
      print(f"{i + 1}. {items[i]}")

def save_items(items, filename):
  with open(filename, "w") as f:
    for item in items:
      f.write(item + "\n")
      print("Saved")

def load_items(filename):
  items = []

  with open(filename, "r") as f:
    for line in f:
      items.append(line.strip())
    f.close()

def main():
  shopping_list = []

  while True:
      print("\nMenu:")
      print("1) Add")
      print("2) Remove")
      print("3) View")
      print("4) Save")
      print("5) Load")
      print("6) Quit")

      choice = input("Choose an option: ")

      if choice == "1":
          item = input("Enter item to add: ")
          add_items(shopping_list, item)

      elif choice == "2":
          item = input("Enter item to remove: ")
          remove_items(shopping_list, item)

      elif choice == "3":
          print_items(shopping_list)

      elif choice == "4":
          filename = input("Enter filename: ")
          save_items(shopping_list, filename)

      elif choice == "5":
          filename = input("Enter filename: ")
          shopping_list = load_items(filename)

      elif choice == "6":
          print("Goodbye!")
          break

      else:
          print("Invalid choice.")


main()
````
